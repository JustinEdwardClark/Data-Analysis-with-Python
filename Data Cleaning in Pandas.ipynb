```python
import pandas as pd
```

```python
call_list = pd.read_excel('Customer Call List.xlsx')
```

```python
call_list
```

```python
call_list = call_list.drop_duplicates()

call_list
```

```python
call_list = call_list.drop(columns = 'Not_Useful_Column')

call_list
```

```python
call_list["Last_Name"] = call_list["Last_Name"].str.strip()
call_list["Last_Name"] = call_list["Last_Name"].str.lstrip("...")
call_list["Last_Name"] = call_list["Last_Name"].str.lstrip("/")
call_list["Last_Name"] = call_list["Last_Name"].str.rstrip("_")

call_list
```

```python
# call_list["Phone_Number"] = call_list["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
# call_list["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# call_list["Phone_Number"] = call_list["Phone_Number"].apply(lambda x: str(x))
# call_list["Phone_Number"] = call_list["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

call_list["Phone_Number"] = call_list["Phone_Number"].str.replace('nan--','')
call_list["Phone_Number"] = call_list["Phone_Number"].str.replace('Na--','')

call_list
```

```python
call_list[["Street Address", "State", "Zip_Code"]] = call_list["Address"].str.split(',',2, expand=True)

call_list
```

```python
call_list["Paying Customer"] = call_list["Paying Customer"].str.replace('Yes','Y')
call_list["Paying Customer"] = call_list["Paying Customer"].str.replace('No','N')

call_list
```

```python
call_list["Do_Not_Contact"] = call_list["Do_Not_Contact"].str.replace('Yes','Y')
call_list["Do_Not_Contact"] = call_list["Do_Not_Contact"].str.replace('No','N')

call_list
```

```python
call_list = call_list.replace('N/a','')
call_list = call_list.replace('NaN','')
call_list = call_list.replace('N-a','')
call_list = call_list.fillna('')

call_list
```

```python
for x in call_list.index:
    if call_list.loc[x, "Do_Not_Contact"] == 'Y':
        call_list.drop(x, inplace=True)
        
call_list        
```

```python
call_list["Do_Not_Contact"].str.replace('','N')
```

```python
for x in call_list.index:
    if call_list.loc[x, "Phone_Number"] == '--':
        call_list.drop(x, inplace=True)
        
call_list
```

```python
call_list["Phone_Number"] = call_list["Phone_Number"].str.replace('|','-')

call_list
```

```python
call_list["Phone_Number"] = call_list["Phone_Number"].str.replace('/','-')

call_list
```

```python
# Reset the index and remove the old index
call_list = call_list.reset_index(drop=True)

# Print the columns of the DataFrame
print(call_list.columns)
```

```python
# Drop the 'Address' column in place
call_list.drop('Address', axis=1, inplace=True)

# Check the modified DataFrame
print(call_list)
```
