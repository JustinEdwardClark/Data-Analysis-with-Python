## Cleaning the Health Camp Detail Dataset

```python
det.head()
```

```python
for column in det:
    print(det[column].value_counts())
```

```python
# Check the datatypes of the Health Camp Detail dataset
print("Health Camp Detail - Data Types:")
print(det.dtypes)
```

```python
# Change datatype of Camp_Start_Date and Camp_End_Date to datetime
det[['Camp_Start_Date', 'Camp_End_Date']] = det[['Camp_Start_Date', 'Camp_End_Date']].apply(pd.to_datetime)

# Check the datatypes of the Health Camp Detail dataset
print("Health Camp Detail - Data Types:")
print(det.dtypes)
```

```python
# Camp_Start_Date and Camp_End_Date by quarter

# Checking Camp Duration
det['Camp_Duration'] = (det['Camp_End_Date'] - det['Camp_Start_Date']).dt.days

# Camp_Start_Date and Camp_End_Date by day
det['Camp_Start_Day'] = det['Camp_Start_Date'].dt.day
det['Camp_End_Day'] = det['Camp_End_Date'].dt.day

# Camp_Start_Date and Camp_End_Date by month
det['Camp_Start_Month'] = det['Camp_Start_Date'].dt.month
det['Camp_End_Month'] = det['Camp_End_Date'].dt.month

det['Camp_Start_Quarter'] = det['Camp_Start_Date'].dt.quarter
det['Camp_End_Quarter'] = det['Camp_End_Date'].dt.quarter

# Filter the dataset by each camp type
camp_type_1 = det[det['Category1'] == 'First']
camp_type_2 = det[det['Category1'] == 'Second']
camp_type_3 = det[det['Category1'] == 'Third']

# Print the results
print("Camp Duration of Camp Type 3:")
print(camp_type_3)
```

## Data Analysis

```python
# Merge camp details with the amount of participants that attended
first_camp_merged = pd.merge(fhca, det, on='Health_Camp_ID')
second_camp_merged = pd.merge(shca, det, on='Health_Camp_ID')
third_camp_merged = pd.merge(thca, det, on='Health_Camp_ID')
```

```python
# Find the engagement of each type of camp
first_camp_engagement = first_camp_merged.groupby('Category1')['Health_Score'].mean()
second_camp_engagement = second_camp_merged.groupby('Category1')['Health_Score'].mean()
third_camp_engagement = third_camp_merged.groupby('Category1')['Number_of_stall_visited'].mean()
```

```python
# Find the attendance of each type of camp
first_camp_attendance = first_camp_merged.groupby('Category1')['Patient_ID'].count()
second_camp_attendance = second_camp_merged.groupby('Category1')['Patient_ID'].count()
third_camp_attendance = third_camp_merged.groupby('Category1')['Patient_ID'].count()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 10))

# Define colors for the camps
colors = ['blue', 'green', 'orange']

# Attendance plot
for i, attendance in enumerate([first_camp_attendance, second_camp_attendance, third_camp_attendance]):
    axes[0].bar(attendance.index, attendance.values, color=colors[i], label=f'Camp {i+1}')

axes[0].set_title('Participant Attendance by Camp Type')
axes[0].set_xlabel('Camp Type')
axes[0].set_ylabel('Attendance')

# Engagement plot
for i, engagement in enumerate([first_camp_engagement, second_camp_engagement, third_camp_engagement]):
    axes[1].bar(engagement.index, engagement.values, color=colors[i], label=f'Camp {i+1}')

axes[1].set_title('Participant Engagement by Camp Type')
axes[1].set_xlabel('Camp Type')
axes[1].set_ylabel('Engagement')

plt.tight_layout()
plt.show()
```

```python
# Find the average duration of each type of camp
duration_of_camp_type = det.groupby('Category1')['Camp_Duration'].mean()
```

```python
# Plot average camp duration vs attendance
colors = ['blue', 'green', 'orange']
plt.bar(duration_of_camp_type.index, duration_of_camp_type.values, color=colors)
plt.title('Average Camp Duration by Camp Type')
plt.xlabel('Camp Type')
plt.ylabel('Duration (Days)')
plt.show()
```

```python
# Merge the datasets on Health_Camp_ID
merged_data = pd.merge(fhca, shca, on='Health_Camp_ID', how='outer')
merged_data = pd.merge(merged_data, thca, on='Health_Camp_ID', how='outer')

# Extract year from Camp_Start_Date and Camp_End_Date columns
merged_data['Year'] = merged_data['Health_Camp_ID'].map(det.set_index('Health_Camp_ID')['Camp_Start_Date'].dt.year)

# Calculate attendance per year
attendance_by_year = merged_data.groupby('Year').agg({'Health_Score_x': 'count', 'Health_Score_y': 'count', 'Number_of_stall_visited': 'sum'})
attendance_by_year = attendance_by_year.rename(columns={'Health_Score_x': 'fhca_attendance', 'Health_Score_y': 'shca_attendance', 'Number_of_stall_visited': 'thca_attendance'})

# Print the attendance per year
print(attendance_by_year)
```

```python
# Create a bar plot of attendance per year
plt.figure(figsize=(10, 6))
plt.bar(attendance_by_year.index, attendance_by_year['fhca_attendance'], label='FHCA')
plt.bar(attendance_by_year.index, attendance_by_year['shca_attendance'], label='SHCA', bottom=attendance_by_year['fhca_attendance'])
plt.bar(attendance_by_year.index, attendance_by_year['thca_attendance'], label='THCA', bottom=attendance_by_year['fhca_attendance'] + attendance_by_year['shca_attendance'])

plt.xlabel('Year')
plt.ylabel('Attendance')
plt.title('Attendance of Health Camps per Year')
plt.legend()

# Set custom x-axis tick values
xtick_values = ['2003', '2004', '2005', '2006']
plt.xticks(attendance_by_year.index, xtick_values)

plt.show()
```

```python
# Create a stacked area plot of attendance per year
plt.figure(figsize=(10, 6))
plt.stackplot(attendance_by_year.index, attendance_by_year['fhca_attendance'], attendance_by_year['shca_attendance'], attendance_by_year['thca_attendance'],
              labels=['FHCA', 'SHCA', 'THCA'], colors=['green', 'red', 'orange'])
plt.xlabel('Year')
plt.ylabel('Attendance')
plt.title('Attendance of Health Camps per Year')
plt.legend()

# Set custom x-axis tick values
xtick_values = ['2003', '2004', '2005', '2006']
plt.xticks(attendance_by_year.index, xtick_values)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a grouped bar plot of attendance per year
plt.figure(figsize=(10, 6))
bar_width = 0.25

years = attendance_by_year.index
fhca_attendance = attendance_by_year['fhca_attendance']
shca_attendance = attendance_by_year['shca_attendance']
thca_attendance = attendance_by_year['thca_attendance']

plt.bar(years - bar_width, fhca_attendance, width=bar_width, label='First Type', color='green')
plt.bar(years, shca_attendance, width=bar_width, label='Second Type', color='red')
plt.bar(years + bar_width, thca_attendance, width=bar_width, label='Third Type', color='orange')

plt.xlabel('Year')
plt.ylabel('Attendance')
plt.title('Attendance of Health Camps per Year')
plt.xticks(years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate the number of registrations
registration_count = len(pro['Patient_ID'].unique())

# Calculate the number of participants in the First, Second, and Third Health Camp Attended datasets
first_attended_count = len(fhca['Patient_ID'].unique())
second_attended_count = len(shca['Patient_ID'].unique())
third_attended_count = len(thca['Patient_ID'].unique())

# Calculate the attendance rates for each camp
first_attendance_rate = (first_attended_count / registration_count) * 100
second_attendance_rate = (second_attended_count / registration_count) * 100
third_attendance_rate = (third_attended_count / registration_count) * 100

# Calculate the drop-off rates for each camp
first_dropoff_rate = 100 - first_attendance_rate
second_dropoff_rate = 100 - second_attendance_rate
third_dropoff_rate = 100 - third_attendance_rate

# Calculate the overall attendance and drop-off rates
overall_attendance_rate = (first_attendance_rate + second_attendance_rate + third_attendance_rate) / 3
overall_dropoff_rate = 100 - overall_attendance_rate

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# First Camp Attendance vs Drop-off Rate
axes[0, 0].pie([first_attendance_rate, first_dropoff_rate], labels=['Attendance', 'Drop-off'], colors=['orange', 'lightblue'], autopct='%1.1f%%', startangle=90)
axes[0, 0].set_title('First Camp Attendance vs Drop-off Rate')

# Second Camp Attendance vs Drop-off Rate
axes[0, 1].pie([second_attendance_rate, second_dropoff_rate], labels=['Attendance', 'Drop-off'], colors=['orange', 'lightblue'], autopct='%1.1f%%', startangle=90)
axes[0, 1].set_title('Second Camp Attendance vs Drop-off Rate')

# Third Camp Attendance vs Drop-off Rate
axes[1, 0].pie([third_attendance_rate, third_dropoff_rate], labels=['Attendance', 'Drop-off'], colors=['orange', 'lightblue'], autopct='%1.1f%%', startangle=90)
axes[1, 0].set_title('Third Camp Attendance vs Drop-off Rate')

# Overall Attendance vs Drop-off Rate
axes[1, 1].pie([overall_attendance_rate, overall_dropoff_rate], labels=['Attendance', 'Drop-off'], colors=['orange', 'lightblue'], autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Overall Attendance vs Drop-off Rate')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the grid of plots
plt.show()
```
